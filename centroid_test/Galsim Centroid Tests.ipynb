```python
import galsim as ga
import numpy as np
import Library as L
import matplotlib.pyplot as plt
import pandas as pd
import deblend
%matplotlib inline
```

```python
flux = 10000; hlr = 2; e1 = 0; e2 = 0; x0 = 0.01; y0 = 0; n = 0.5
x_len = y_len = 100; pixel_scale = 0.2
method = 'fft'
```

```python
gal = L.create_galaxy(flux,hlr,e1,e2,x0,y0,return_obj=True,method=method,x_len=x_len,y_len=y_len,scale=pixel_scale)
```

```python
gal.centroid()
```

```python
image = ga.ImageD(x_len, y_len, scale=pixel_scale)
im = gal.drawImage(image=image, method=method)
```

```python
# This returns the value counting from a one-indexed system
im.center()
```

```python
im.bounds
```

```python
plt.imshow(im.array,interpolation='none',origin='lower')
```

```python
# Now check the rotation peak
rot_image, rot_peak = deblend.rotate(im.array,[x0/pixel_scale,y0/pixel_scale])
```

```python
rot_peak
```

```python
# This will return the center counting from a zero-indexed system. 
y,x = np.where(image.array == np.max(image.array))
```

```python
num = 100
center_counts = np.zeros((num,))
x_arr = np.linspace(-0.02,0.02,num)
for i in xrange(0,num):
    x0 = x_arr[i]
    gal = L.create_galaxy(flux,hlr,e1,e2,x0,y0,return_obj=True,method=method,x_len=x_len,y_len=y_len,scale=pixel_scale)
    im = gal.drawImage(image=image, method=method)
    x,y = np.where(image.array == np.max(image.array))
    # Count the number of pixels with the max value in the image. 
    center_counts[i] = len(x)

fig = plt.figure(figsize=(15,11))
fs = 20
plt.scatter(x_arr,center_counts)
plt.xlabel('x Coordinate of Centroid',fontsize=fs); plt.ylabel('Pixels',fontsize=fs)
plt.title('Observed Number of Pixels with Maximum Values For Even Sized Image',fontsize=fs)
```
