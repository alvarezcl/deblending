```python
## This script creates two objects, blends, deblends and then measures
## the shape of the deblended objects and compares to the original. 
## It now incorporates simultaneous fitting.

from __future__ import division
import Library
import galsim
import deblend
import sys
import numpy as np
import lmfit
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import triangle
import os
%matplotlib inline
```

```python
# Simple function that runs a loop over the main functions and returns key information regarding fits.    
def run_sample_batch(num_trials_deblender, num_trials_simult_fitter, run_simult_fit=False):
    
    # Galsim function definitions
    sersic_func = galsim.Sersic
    
    # Set the RNG
    seed_int = 10
    seed_1 = galsim.BaseDeviate(seed_int)
    seed_2 = galsim.BaseDeviate(seed_int**2)
    seed_3 = galsim.BaseDeviate(seed_int**3)
    
    # Image properties
    pixel_scale = 1/5     # arcsec / pixel
    x_len = y_len = 100            # pixel
    image_params = pd.Series(np.array([pixel_scale,x_len,y_len]),index=['pixel_scale','x_len','y_len'])
    
    # Parameters for object a
    flux_a = 25000 # counts
    hlr_a = 1  # arcsec
    e1_a = 0.0 
    e2_a = 0.0
    x0_a_pix = -5; x0_a = x0_a_pix*pixel_scale  # arcsec
    y0_a_pix = 0; y0_a = y0_a_pix*pixel_scale  # arcsec

    n_a = 0.5
    
    # Parameters for object b
    flux_b = 25000 # counts
    hlr_b = hlr_a # arcsec
    e1_b = 0.0
    e2_b = 0.0
    x0_b_pix = 5; x0_b = x0_b_pix*pixel_scale  # arcsec
    y0_b_pix = 0; y0_b = y0_b_pix*pixel_scale  # arcsec
    n_b = 0.5
    
    truth = pd.Series(np.array([flux_a,hlr_a,e1_a,e2_a,x0_a,y0_a,
                                flux_b,hlr_b,e1_b,e2_b,x0_b,y0_b]),
                                index=['flux_a','hlr_a','e1_a','e2_a','x0_a','y0_a',
                                       'flux_b','hlr_b','e1_b','e2_b','x0_b','y0_b'])
    
    # Use LSST defined sky noise for r-band
    add_noise_flag = True
    texp = 6900 # seconds;
    sbar = 26.8 # sky photons per second per pixel
    sky_level = 0 # For now
    sky_noise = np.sqrt(sky_level)
    texp = 0; # For only poisson noise
    
    # psf properties
    psf_flag = False
    beta = 3
    fwhm_psf = 0.6
    
    # FFT
    method = 'fft'
    
    # Store the results of the fits to the deblended children
    results_deblend = []
    results_true = []
    images = []
    divisor = 100
    factor_init = 1
    separations = []
    for i in xrange(0,num_trials_deblender):
        # Random offsetting for image
        np.random.RandomState(seed=1)
        x_offset_a = np.random.uniform(-pixel_scale/2,pixel_scale/2) # interval of 1/2 pixel
        x_offset_b = np.random.uniform(-pixel_scale/2,pixel_scale/2) # interval of 1/2 pixel
        x0_a_eff = x0_a + x_offset_a
        x0_b_eff = x0_b + x_offset_b
        sep = x0_b_eff - x0_a_eff
        separations.append(sep)
        results_deb, results_tr, children = Library.deblend_estimate(flux_a,hlr_a,e1_a,e2_a,x0_a_eff,y0_a,n_a,
                                                                     flux_b,hlr_b,e1_b,e2_b,x0_b_eff,y0_b,n_b,
                                                                     sersic_func, seed_1, seed_2, seed_3, 
                                                                     pixel_scale, x_len, y_len, 
                                                                     add_noise_flag, texp, sbar, 
                                                                     psf_flag, beta, fwhm_psf,
                                                                     method,
                                                                     factor_init,
                                                                     plot=False)
        results_deblend.append(results_deb)
        results_true.append(results_tr)
        if np.mod(i,divisor) == 0:
            images.append([children,i])
    separations = pd.Series(separations)
    results_deblend = pd.DataFrame(results_deblend)
    results_true = pd.DataFrame(results_true)

    
    if run_simult_fit == True:
        # Run the simultaneous fitter
        results_sim = []
        for i in xrange(0,num_trials_simult_fitter):
            # Random offsetting for image
            x_offset_a = np.random.uniform(-pixel_scale/2,pixel_scale/2) # interval of 1/2 pixel
            x_offset_b = np.random.uniform(-pixel_scale/2,pixel_scale/2) # interval of 1/2 pixel
            x0_a_eff = x0_a + x_offset_a
            x0_b_eff = x0_b + x_offset_b
            im_no_noise, im_noise, lm_results = Library.run_2_galaxy_full_params_simple(flux_a,hlr_a,e1_a,e2_a,x0_a_eff,y0_a,n_a,
                                                                                        flux_b,hlr_b,e1_b,e2_b,x0_b_eff,y0_b,n_b,
                                                                                        psf_flag,beta,fwhm_psf,
                                                                                        x_len,y_len,pixel_scale,sersic_func,sersic_func,seed_1,seed_2,seed_3,
                                                                                        add_noise_flag,sky_level)    
                                                                                                  
            results_sim.append(Library.rearrange_lmfit_2obj(lm_results))

        results_sim = pd.DataFrame(results_sim)
        return results_deblend, results_true, results_sim, truth, seed_int, image_params, images, separations

    
    return results_deblend, results_true, truth, seed_int, image_params, images, separations
```

```python
# Run a number of trials for bias analysis for deblending and simult fitting
deblend_runs = 1000; simult_runs = 1000; 
results_deblend, results_true, results_sim, truth, seed_int, image_params, dbl_im, separations = run_sample_batch(deblend_runs,simult_runs,run_simult_fit=True)
```

# We now show the triangle plot of the results of single fits to the child objects.

```python
figure_1 = triangle.corner(results_deblend,labels=results_deblend.columns,truths=truth,
                             show_titles=True, title_args={'fontsize':20})
figure_1.gca().annotate('Results of Single Fits to Deblended Children', xy=(0.5,1.0), xycoords='figure fraction')
```

```python
# Access a random set of data for the best fit and compare, access a random index and plot the deblended with fits
index = int(len(dbl_im)*np.random.random(1))
dbl_obj_im_a = dbl_im[index][0][0]
dbl_obj_im_b = dbl_im[index][0][1]
index_val = dbl_im[index][1]
fit_to_dbl_obj = results_deblend.ix[index_val]
```

```python
# Create the true images of the unblended profiles
true_im_a = Library.create_galaxy(truth['flux_a'],truth['hlr_a'],truth['e1_a'],truth['e2_a'],truth['x0_a'],truth['y0_a'],
                                  x_len=image_params['x_len'],y_len=image_params['y_len'],scale=image_params['pixel_scale'])
true_im_b = Library.create_galaxy(truth['flux_b'],truth['hlr_b'],truth['e1_b'],truth['e2_b'],truth['x0_b'],truth['y0_b'],
                                  x_len=image_params['x_len'],y_len=image_params['y_len'],scale=image_params['pixel_scale'])

# Create the images of the fits to the deblended objects.x_len=image_params['x_len'],y_len=image_params['y_len'],scale=image_params['pixel_scale'])
fit_dbl_a = Library.create_galaxy(fit_to_dbl_obj['flux_a'],fit_to_dbl_obj['hlr_a'],fit_to_dbl_obj['e1_a'],fit_to_dbl_obj['e2_a'],fit_to_dbl_obj['x0_a'],fit_to_dbl_obj['y0_a'],
                                  x_len=image_params['x_len'],y_len=image_params['y_len'],scale=image_params['pixel_scale'])
fit_dbl_b = Library.create_galaxy(fit_to_dbl_obj['flux_b'],fit_to_dbl_obj['hlr_b'],fit_to_dbl_obj['e1_b'],fit_to_dbl_obj['e2_b'],fit_to_dbl_obj['x0_b'],fit_to_dbl_obj['y0_b'],
                                  x_len=image_params['x_len'],y_len=image_params['y_len'],scale=image_params['pixel_scale'])
```

# We compare the following: true un-blended objects, deblended objects, and their fits by inspecting images. 

```python
# Plot the fits to the deblended profiles vs the deblended proflies vs the true profiles
gs = gridspec.GridSpec(7,9)                                   
fig = plt.figure(figsize=(15,25))
sh = 0.8
plt.suptitle('  True Objects vs Deblended Objects vs Fits to Deblended Objects\n For Separation: '+ str(truth['x0_b']-truth['x0_a']) + '\"',fontsize=20)

# Plot the true blend
ax = fig.add_subplot(gs[0,1:7])
z = ax.imshow(true_im_a.array + true_im_b.array,interpolation='none',origin='lower'); plt.title('True Blend'); plt.colorbar(z,shrink=sh)

# Plot the true objects
ax1 = fig.add_subplot(gs[1,0:4])
a = ax1.imshow(true_im_a.array,interpolation='none',origin='lower'); plt.title('True Object A'); plt.colorbar(a,shrink=sh)
ax2 = fig.add_subplot(gs[1,6:10])
b = ax2.imshow(true_im_b.array,interpolation='none',origin='lower'); plt.title('True Object B'); plt.colorbar(b,shrink=sh)

# Plot the deblended objects
ax3 = fig.add_subplot(gs[2,0:4])
c = ax3.imshow(dbl_obj_im_a,interpolation='none',origin='lower'); plt.title('Deblended Object A'); plt.colorbar(c,shrink=sh)
ax4 = fig.add_subplot(gs[2,6:10])
d = ax4.imshow(dbl_obj_im_b,interpolation='none',origin='lower'); plt.title('Deblended Object B'); plt.colorbar(d,shrink=sh)

# Plot the residual of the deblended and the true individual objects
ax5 = fig.add_subplot(gs[3,0:4])
e = ax5.imshow(dbl_obj_im_a-true_im_a.array,interpolation='none',origin='lower'); plt.title('Residual of Deblended A and True A'); plt.colorbar(e,shrink=sh)
ax6 = fig.add_subplot(gs[3,6:10])
f = ax6.imshow(dbl_obj_im_b-true_im_b.array,interpolation='none',origin='lower'); plt.title('Residual of Deblended B and True B'); plt.colorbar(f,shrink=sh)

# Plot the fits to the deblended objects
ax7 = fig.add_subplot(gs[4,0:4])
g = ax7.imshow(fit_dbl_a.array,interpolation='none',origin='lower'); plt.title('Fit To Deblended Object A'); plt.colorbar(g,shrink=sh)
ax8 = fig.add_subplot(gs[4,6:10])
h = ax8.imshow(fit_dbl_b.array,interpolation='none',origin='lower'); plt.title('Fit To Deblended Object B'); plt.colorbar(h,shrink=sh)

# Plot the residual of the fit to the deblended to the deblended objects
ax9 = fig.add_subplot(gs[5,0:4])
i = ax9.imshow(fit_dbl_a.array-dbl_obj_im_a,interpolation='none',origin='lower'); plt.title('Residual Of Fit To Deblended Object A and Deblended A'); plt.colorbar(i,shrink=sh)
ax10 = fig.add_subplot(gs[5,6:10])
j = ax10.imshow(fit_dbl_b.array-dbl_obj_im_b,interpolation='none',origin='lower'); plt.title('Residual Of Fit To Deblended Object B and Deblended B'); plt.colorbar(j,shrink=sh)

# Plot the residual of the fit to the deblended to the true unblended object
ax11 = fig.add_subplot(gs[6,0:4])
k = ax11.imshow(fit_dbl_a.array-true_im_a.array,interpolation='none',origin='lower'); plt.title('Residual Of Fit To Deblended Object A and True Object A'); plt.colorbar(k,shrink=sh)
ax12 = fig.add_subplot(gs[6,6:10])
l = ax12.imshow(fit_dbl_b.array-true_im_b.array,interpolation='none',origin='lower'); plt.title('Residual Of Fit To Deblended Object B and True Object B'); plt.colorbar(l,shrink=sh)

plt.show()
```
