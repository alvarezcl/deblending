# Extract Features

```python
# Filter out noisy peaks
def filter_mini_peaks(im,limit):
    labeled, n_nuclei = mh.label(im)
    #pdb.set_trace()
    tot_region = np.zeros(im.shape)
    for i in xrange(1,n_nuclei+2):
        labeled_i = np.copy(labeled)
        labeled_i[labeled_i != i] = 0
        if (np.sum(labeled_i) < limit):
            labeled_i[labeled_i == i] = 0
        reg_i = labeled_i*np.copy(im)
        tot_region = tot_region + reg_i
    return tot_region
new_data = filter_mini_peaks(np.copy(filtered_data),100)
plt.imshow(new_data,interpolation ='bilinear',origin='lower',cmap = plt.get_cmap(name='hot'))
plt.colorbar()
```

```python
# Asym
a_rms, a_abs, a_abs_mat, a_est, peak = cs.compute_asym(new_data,[0,0],x_len,y_len)
# M20
m20, m20mat, peak = cs.find_min_m20(new_data,x_len,y_len)
# Gini coefficient
gini = cs.compute_gini(new_data)
gini
```

```python
# Use HOG
from skimage.feature import hog
from skimage import color, exposure
fd, hog_image = hog(new_data, orientations=8, pixels_per_cell=(2, 2),
                    cells_per_block=(2,2), visualise=True, normalise=True)
```

```python
# Segmentation for Peak Finding
def find_peaks(im,interval):
    ranges = np.linspace(im.min(),im.max(),interval)
    tot_nuclei = []
    labels_arr = []
    matrix = im
    for val in ranges:
        matrix[matrix < val] = 0
        new_matrix = filter_mini_peaks(np.copy(matrix),2)
        labels_i, n_nuclei_i = mh.label(matrix)
        tot_nuclei.append(n_nuclei_i)
        labels_arr.append(labels_i)
    return pd.Series(tot_nuclei), pd.Panel(labels_arr)
```

```python
regions, labels = find_peaks(np.copy(new_data),10000)
```

```python
if (max(regions ==2)):
    print 'True'
    where = np.where(regions==2)
    where_1st_2 = where[0][0]
    first_label = labels[where_1st_2]
    two_peaks = first_label*new_data
```

```python
# Watershed
from skimage import morphology
from scipy import ndimage
from skimage.feature import peak_local_max
image = new_data > np.sqrt(sky_level)
copy = np.copy(new_data)
# Now we want to separate the two objects in image
# Generate the markers as local maxima of the distance to the background
distance = ndimage.distance_transform_edt(image)
local_maxi = peak_local_max(copy, indices=False, footprint=np.ones((100, 100)),
                            labels=image)
markers = ndimage.label(local_maxi)[0]
labels = morphology.watershed(-copy, markers, mask=copy)
plt.imshow(labels,interpolation='none',origin='lower',cmap=plt.cm.hot)
plt.colorbar()
```
